```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
#reg-regression plot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df,hue='size')
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df,hue='size',palette="husl")
plt.show()
```

```python
#kde-kernal distributive element
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df,kind='reg')
plt.show()
```

```python
#controlling the markers
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
sb.pairplot(df,hue='sex',markers=['o','s'])
plt.show()
```

```python
#advanced customization with seaborn pairplot
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset("tips")
g=sb.pairplot(df,hue='day')
g.fig.suptitle("Pair of TIPS Dataset",y=1.02)
g.set(xticks=[],yticks=[])#remove tick labels
plt.show()
```
